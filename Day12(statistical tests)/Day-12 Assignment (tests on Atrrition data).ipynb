```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(url)
pd.set_option('display.max_columns',None)
df.head()
```

```python
df_Yes_ATTR=df[df['Attrition'] == 'Yes'] # 1 independent sample 
df_No_ATTR=df[df.Attrition == 'No']    # 2nd independent sample
```

```python
df_Yes_ATTR.head(2)
```

```python
df_No_ATTR.head(2)
```

# Mann whitney Test -> To compare 2 independent samples

```python
'''
H0 : there is no significant diffrence in the monthly income of Attrition_Yes and Attrition_No
H1 :there is significant diffrence in the monthly income of Attrition_Yes and Attrition_No

here p_value >= 0.05  = so we accept H0 
'''

from scipy.stats import mannwhitneyu
_, p_value=mannwhitneyu(df_Yes_ATTR.MonthlyIncome ,df_No_ATTR.MonthlyIncome)
p_value
```

```python
from scipy.stats import chi2_contingency
# 1 we need to perform cross tabulation 
```

```python
# for Attrition and BusinessTravel
chitable=pd.crosstab(df.Attrition,df.BusinessTravel)
chitable
```

```python
chi2, p, dof, ex =chi2_contingency(chitable)
'''
H0 : there is no dependecy between Attrition and BusinessTravel
H1 : there is dependecy between Attrition and BusinessTravel
p<0.05 = so reject the H0 
'''
p
```

```python
# Atrrition and Department
chitable1=pd.crosstab(df.Attrition,df.Department)
chitable1
```

```python
chi2, p, dof, ex =chi2_contingency(chitable1)
'''
H0 : there is no dependecy between Attrition and BusinessTravel
H1 : there is dependecy between Attrition and BusinessTravel
p<0.05 = so reject the H0 
'''
p
```
